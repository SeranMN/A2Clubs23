```python
import os
from PIL import Image

import cv2
def compressImg(path):
    print(path, end = "==========")
    img = Image.open(path)
    originalSize = img.size
    sizeB = os.stat(path).st_size//1024
    print(os.stat(path).st_size//1024,"KB", end = "-->")
    if(sizeB>300):
        img.thumbnail((250,250))
        img = img.transform(originalSize, Image.EXTENT, (0,0, 250, 250))
        img.save(path, quality=10, optimize=True)
    else:
        img.save(path, quality=50, optimize=True)
    print(os.stat(path).st_size//1024,"KB")


# compressImg('./test.png')
```

```python
for path, currentDirectory, files in os.walk(os.getcwd()):
    for file in files:
        if(file.endswith('.png') or file.endswith('.jpg') or file.endswith('.JPG')  or file.endswith('.jpeg')):
            filePath = os.path.join(path, file)
            try:
                compressImg(filePath)
            except:
                print("==============FAILED======")
            
```
